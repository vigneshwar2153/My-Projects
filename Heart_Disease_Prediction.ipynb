```python
data = pd.read_csv('heart_disease.csv')
data.dropna(inplace=True)
data.drop(columns=['education'], inplace=True)
```

```python
print(data)
```

```python
data['prevalentStroke'] = data['prevalentStroke'].replace(['yes','no'],[1,0])
```

```python
data['Heart_ stroke'] = data['Heart_ stroke'].replace(['yes','No'],[1,0])
```

```python
print(data)
```

```python
data_input = data.drop(columns=['Heart_ stroke'])
data_output = data['Heart_ stroke']
```

```python
data_output.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=2)
```

```python
data_input,data_output = smote.fit_resample(data_input,data_output.ravel())
```

```python
import numpy as np
np.unique(data_output,return_counts=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_tain,x_test,y_train,y_test = train_test_split(data_input,data_output,test_size=0.2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_tain,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
coef = model.coef_[0]
features = x_test.columns  # if using DataFrame
pd.Series(coef, index=features).sort_values().plot(kind='barh')
plt.title('Feature Importance (Logistic Regression Coefficients)')
plt.show()
```
